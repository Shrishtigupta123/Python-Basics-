14.Describe a scenario where a while loop is more suitable than a for loop?
Suppose you are building a game where the player has to guesss a random number between 1 to 100.The game should keep asking the player for their guess until they correctly guess the number.
In this scenario , a While loop is more suitable than a for loop for several reasons:
1)Unknown no of iterations: The game doesn't know how many attempts the player will need to correctly guess the number A while loop allows the game to continue asking for guesses until the correct ans is given .
2) Conditional loooping: The game needs to continue looping until a certain condition is met i.e the player correctly guesses the number. A While loop is designed for conditional looping, making it a natural fit for this scenario.
3)No fixed sequnce unlike for loop, which iterates over a fixed sequence, a while loop allows the game to dynamically adjust the number of iterations based on the player's progress.

```python
#Q6.Write a function to demonstrate the behavior of mutable and immutable arguments?
def modify_immutable(num):
    num += 1
    print(f"Inside function: num = {num}")

def modify_mutable(my_list):
    my_list.append(4)
    print(f"Inside function: my_list = {my_list}")

original_num = 5
print(f"Before function call: original_num = {original_num}")
modify_immutable(original_num)
print(f"After function call: original_num = {original_num}")

my_list = [1, 2, 3]
print(f"Before function call: my_list = {my_list}")
modify_mutable(my_list)
print(f"After function call: my_list = {my_list}")

```

```python
#Q8. Write a program to demonstrate the use of logical operators?
a = True
b = False
print("AND Operator")
print(f"{a and b}") #Output is False because one of them is False.
print(f"{a and a}") #Output will be True only when both of them are True
print("")
print("OR Operator")
print(f"{a or b}") #Output is True because one of them is True.
print(f"{b or b}") #Output will be False only when both of them are False.
print("")
print("NOT Operator")
print(f"{not a}") #Output is False because a is True.
print(f"{not b}") #Output is True because b is False.
```
